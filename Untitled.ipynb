```python
import time
size=10000
alist = list(range(size))
t = time.perf_counter()
asum = sum(alist)
print("sum(alist) : %f sec" % (time.perf_counter() - t))

x=np.array(alist)
t=time.perf_counter()
xsum1=x.sum()
print("npa.sum() : %f sec" % (time.perf_counter() - t))


t=time.perf_counter()
xsum2=np.sum(x)
print("np.sum(npa) : %f sec" % (time.perf_counter() - t))
print(asum,xsum1,xsum2)








```
